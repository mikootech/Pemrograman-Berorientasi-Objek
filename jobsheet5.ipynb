## Praktikum 01

```python
# Kelas Induk
class Burung:
    def __init__(self, nama):
        self.nama = nama

    def terbang(self):
        print(f"{self.nama} terbang dengan cara umum.")

    def bersuara(self):
        print(f"{self.nama} mengeluarkan suara burung.")

# Kelas Anak 1
class Elang(Burung):
    def __init__(self, nama, rentang_sayap):
        super().__init__(nama)
        self.rentang_sayap = rentang_sayap

    def terbang(self):
        print(f"{self.nama} terbang tinggi melayang di angkasa.")

    def bersuara(self):
        print(f"{self.nama} berteriak nyaring!")

# Kelas Anak 2
class Pipit(Burung):
    def __init__(self, nama, warna_bulu):
        super().__init__(nama)
        self.warna_bulu = warna_bulu

    def terbang(self):
        print(f"{self.nama} terbang cepat di antara pepohonan.")

    def bersuara(self):
        print(f"{self.nama} berkicau merdu: Cit cit!")

# Fungsi Polimorfik
def demonstrasi_aksi_burung(daftar_burung):
    print("\nAksi Burung:")
    for burung in daftar_burung:
        print(f"-- Aksi untuk {burung.nama} --")
        burung.terbang()
        burung.bersuara()
        print("-" * 15)

if __name__ == "__main__":
    elang_jawa = Elang("Elang Jawa", 1.5)
    pipit_gereja = Pipit("Pipit Gereja", "Coklat")
    burung_aneh = Burung("Burung Misterius")
    
    koleksi_burung = [elang_jawa, pipit_gereja, burung_aneh]
    demonstrasi_aksi_burung(koleksi_burung)
```

## Praktikum 02

```python
class Kucing:
    def bersuara(self):
        print("Kucing: Meow!")

class Anjing:
    def bersuara(self):
        print("Anjing: Guk guk!")

class Bebek:
    def bersuara(self):
        print("Bebek: Kwek kwek!")

class Mobil:
    def jalankan(self):
        print("Mobil: Brummm!")

def buat_suara(objek):
    try:
        objek.bersuara()
    except AttributeError:
        print(f"Objek {type(objek).__name__} tidak bisa bersuara.")

if __name__ == "__main__":
    daftar_objek = [Kucing(), Anjing(), Bebek(), Mobil()]
    print("Demonstrasi Duck Typing:")
    for item in daftar_objek:
        buat_suara(item)
```

## Praktikum 03

```python
class KalkulatorSederhana:
    def __init__(self, nama="Kalkulator"):
        self.nama = nama
        print(f"{self.nama} siap digunakan.")

    def tambah(self, *args):
        print(f"\nMemanggil metode tambah dengan argumen: {args}")
        if not args:
            print("Tidak ada angka untuk dijumlahkan.")
            return 0
        
        total = 0
        valid_input = True
        for angka in args:
            if isinstance(angka, (int, float)):
                total += angka
            else:
                print(f"Peringatan: Argumen '{angka}' bukan angka.")
                valid_input = False
        
        print(f"Hasil penjumlahan: {total}")
        return total

if __name__ == "__main__":
    calc = KalkulatorSederhana("Calc-01")
    calc.tambah(5, 10)
    calc.tambah(2, 3, 5, 10)
    calc.tambah(1, 2, "tiga", 4)
```

## Praktikum 04

```python
class DataProcessor:
    def __init__(self, processor_id):
        self.processor_id = processor_id
        print(f"Data Processor {self.processor_id} siap.")

    def process(self, *args, **kwargs):
        print(f"\n--- {self.processor_id} Memproses Data ---")
        if args:
            print("Argumen Posisi Diterima:")
            for i, arg in enumerate(args):
                print(f"  args[{i}]: {arg}")
        
        if kwargs:
            print("Argumen Kata Kunci Diterima:")
            for kunci, nilai in kwargs.items():
                print(f"  {kunci}: {nilai}")

if __name__ == "__main__":
    p1 = DataProcessor("DP-001")
    p1.process(100, "OK")
    p1.process(user="admin", level=5)
    p1.process("Task-A", status="Running")
```

## Praktikum 05

```python
import math

class Bentuk:
    def nama_bentuk(self):
        return "Bentuk Generik"
    def hitung_luas(self):
        raise NotImplementedError("Subclass harus implementasi ini")

class Lingkaran(Bentuk):
    def __init__(self, radius):
        self.radius = radius
    def nama_bentuk(self):
        return "Lingkaran"
    def hitung_luas(self):
        return math.pi * (self.radius ** 2)

class Persegi(Bentuk):
    def __init__(self, sisi):
        self.sisi = sisi
    def nama_bentuk(self):
        return "Persegi"
    def hitung_luas(self):
        return self.sisi * self.sisi

class TaplakMeja: # Tidak ikut inheritance
    def __init__(self, p, l):
        self.p, self.l = p, l
    def hitung_luas(self):
        return self.p * self.l

def tampilkan_info_luas(objek):
    print("-" * 20)
    try:
        luas = objek.hitung_luas()
        try:
            nama = objek.nama_bentuk()
        except AttributeError:
            nama = type(objek).__name__
        print(f"Objek: {nama}\nLuas: {luas:.2f}")
    except Exception as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    items = [Lingkaran(7), Persegi(5), TaplakMeja(1.5, 0.8), Bentuk()]
    for item in items:
        tampilkan_info_luas(item)
```

## Praktikum 06

```python
class Burung:
    def __init__(self, nama): self.nama = nama
    def terbang(self): print(f"{self.nama} terbang.")

class Elang(Burung):
    def berburu(self): print(f"{self.nama} sedang berburu.")

class Pipit(Burung):
    def bangun_sarang(self): print(f"{self.nama} membangun sarang.")

def interaksi(burung):
    if isinstance(burung, Burung):
        burung.terbang()
        if isinstance(burung, Elang):
            burung.berburu()
        elif isinstance(burung, Pipit):
            burung.bangun_sarang()
    else:
        print("Bukan Burung!")

if __name__ == "__main__":
    interaksi(Elang("Elang Sumatra"))
    interaksi(Pipit("Pipit Rumah"))
```

## Program 01

##### Hierarki Komputer

```python
class Komputer:
    def __init__(self, merk, processor, ram_gb):
        self.merk = merk
        self.processor = processor
        self.ram_gb = ram_gb

    def info_spec(self):
        print(f"Merk: {self.merk}")
        print(f"Processor: {self.processor}")
        print(f"RAM: {self.ram_gb} GB")

    def install_software(self, *args): # Simulasi Overloading (Opsi B)
        if not args:
            print(f"Tidak ada software baru di {self.merk}.")
        else:
            for sw in args:
                print(f"Menginstall {sw} di {self.merk}...")

class Laptop(Komputer):
    def __init__(self, merk, proc, ram, layar, berat):
        super().__init__(merk, proc, ram)
        self.layar = layar
        self.berat = berat

    def info_spec(self): # Overriding
        super().info_spec()
        print(f"Ukuran Layar: {self.layar} inch")
        print(f"Berat: {self.berat} Kg")

class Desktop(Komputer):
    def __init__(self, merk, proc, ram, casing, monitor_ext):
        super().__init__(merk, proc, ram)
        self.casing = casing
        self.monitor_ext = monitor_ext

    def info_spec(self): # Overriding
        super().info_spec()
        print(f"Jenis Casing: {self.casing}")
        print(f"Monitor External: {'Ya' if self.monitor_ext else 'Tidak'}")

def cetak_semua_spesifikasi(daftar_komputer):
    print("\n--- DAFTAR SPESIFIKASI KOMPUTER ---")
    for comp in daftar_komputer:
        comp.info_spec()
        print("-" * 25)

if __name__ == "__main__":
    my_laptop = Laptop("ROG Strix", "Ryzen 9", 32, 15.6, 2.3)
    my_pc = Desktop("Custom Build", "Core i9", 64, "Tower", True)
    
    list_komp = [my_laptop, my_pc]
    cetak_semua_spesifikasi(list_komp)
    
    # Demonstrasi Overloading
    print("Simulasi Overloading:")
    my_laptop.install_software("VS Code", "Docker", "Git")
```
